```python
import h5py
import matplotlib.pyplot as plt
```

```python
n = 500
```

```python
def plot(filename, title = "Signal"):
    f = h5py.File(filename, 'r')
    print("Keys: %s" % f.keys())
    data = f["Raw"]["Reads"]
    data = f["Raw"]["Reads"]["Read_981"]["Signal"]
    print(data)
    plt.figure(figsize=(12,6))
    plt.title(title, fontsize=20)
    plt.plot(range(n), data[:n])
```

```python
filename = "./data_ref/data/references/ecoli_reference.fast5"
plot(filename, "Ecoli reference")
```
